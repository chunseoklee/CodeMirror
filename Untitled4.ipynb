Now, let's load the Whisper model. For a 'turbo' like experience, you might consider smaller models like `openai/whisper-tiny.en` for English, or `openai/whisper-small` for multilingual. I'll use `openai/whisper-tiny.en` as an example for its speed. If you need a different language or higher accuracy, you can replace it with a larger model like `openai/whisper-base` or `openai/whisper-medium`.

```python
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline
from datasets import load_dataset


device = "cuda:0" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32

model_id = "openai/whisper-large-v3-turbo"

model = AutoModelForSpeechSeq2Seq.from_pretrained(
    model_id, torch_dtype=torch_dtype, low_cpu_mem_usage=True, use_safetensors=True
)
model.to(device)

processor = AutoProcessor.from_pretrained(model_id)

pipe = pipeline(
    "automatic-speech-recognition",
    model=model,
    tokenizer=processor.tokenizer,
    feature_extractor=processor.feature_extractor,
    torch_dtype=torch_dtype,
    device=device,
)
model
#dataset = load_dataset("distil-whisper/librispeech_long", "clean", split="validation")
#sample = dataset[0]["audio"]

#result = pipe(sample)
#print(result["text"])
```

To observe the input and output of each layer, we'll use PyTorch's forward hooks. First, we define a dictionary to store the activations and a hook function that saves the input and output of each module it's attached to.

```python
activations = {}
def save_activation(name):
    def hook(module, input, output):
        # Debug print: confirm if the hook is being triggered
        # print(f"Hook triggered for layer: {name}")
        activations[name] = {'input': input, 'output': output}
    return hook
```

Now, we iterate through the entire model and register this hook for each submodule. We'll focus on the encoder and decoder layers as they contain the core computations.

```python
for name, module in model.named_modules():
    # You can be more selective here, e.g., if 'encoder.layers' in name or 'decoder.layers' in name
    # For this example, let's just look at specific sub-layers to keep the output manageable
    if isinstance(module, (torch.nn.Linear, torch.nn.Conv1d, torch.nn.LayerNorm)) and 'encoder' in name:
        module.register_forward_hook(save_activation(name))
```

Next, we need a sample audio input to pass through the model to trigger these hooks. We'll use the `load_dataset` function from `datasets` to get an audio sample, similar to how it's done in the commented-out code in your notebook.

```python
sample = load_dataset("audiofolder", data_files="robot.mp3", split="train")
# The dataset object usually contains a list of samples. We need the first one.
sample_audio = sample[0]["audio"]

result = pipe(sample_audio, generate_kwargs={"language": "en"})
print(result["text"])
```

Finally, we can inspect the `activations` dictionary to see the shapes of the inputs and outputs for each registered layer.

```python
print("--- Activations (Input/Output Shapes) ---")
for name, data in activations.items():
    input_shape = [i.shape for i in data['input']] if isinstance(data['input'], (list, tuple)) else data['input'].shape
    output_shape = [o.shape for o in data['output']] if isinstance(data['output'], (list, tuple)) else data['output'].shape
    print(f"Layer: {name}")
    print(f"  Input shape(s): {input_shape}")
    print(f"  Output shape(s): {output_shape}")
    print("----------------------------------------")
```
